### Part 2 - Practice with Grabbing Data

Let's now get some practice with grabbing certain parts of the data. If you'd like some extra practice, try answering each of the questions in more than one way (because remember, we can often grab our data in a couple of different ways). 

1. Grab the first 10 rows of the `chlorides` column. 
2. Grab the last 10 rows of the `chlorides` column. 
3. Grab indices 264-282 of the `chlorides` **and** `density` columns. 
4. Grab all rows where the `chlorides` value is less than 0.10. 
5. Now grab all the rows where the `chlorides` value is greater than the column's mean (try **not** to use a hard-coded value for the mean, but instead a method).
6. Grab all those rows where the `pH` is greater than 3.0 and less than 3.5. 
7. Further filter the results from 6 to grab only those rows that have a `residual sugar` less than 2.0. 

```python
# 1. Grab the first 10 rows of the chlorides column. 
df['chlorides'].head(10)
df.chlorides.head(10)
df['chlorides'][:10]
df.chlorides[:10]
df.iloc[:10, 4]          # cave: rows 0 to 10 (excluded) with iloc
df.loc[:9, 'chlorides']  # cave: rows 0 to 9 with loc
```

```python
# 2. Grab the last 10 rows of the chlorides column. 
df['chlorides'].tail(10)
df.chlorides.tail(10)
df['chlorides'][-10:]
df.chlorides[-10:]
df['chlorides'][4888:]
df.chlorides[4888:]
df.iloc[-10:, 4]           
df.loc[4888:, 'chlorides']
```

```python
# 3. Grab indices 264-282 of the chlorides and density columns. 
df[['chlorides', 'density']][264:283]      # Cave: 283 (not included)
df.loc[264:282, ['chlorides', 'density']]  # Cave: 282 (included) - since it is an explicitely labeled location!
df.iloc[264:283, [4,7]]                    # Cave: 283 (not included)
```

```python
# 4. Grab all rows where the chlorides value is less than 0.10.
df.query('chlorides < 0.10')
df[df.chlorides < 0.10]
df[df['chlorides'] < 0.10]
```

```python
# 5. Now grab all the rows where the chlorides value is greater than the column's mean 
# (try not to use a hard-coded value for the mean, but instead a method).
df[df.chlorides > df.chlorides.mean()]   # chlorides mean is 0.0457723560636995
df[df['chlorides'] > df['chlorides'].mean()]
```

```python
# 6. Grab all those rows where the pH is greater than 3.0 and less than 3.5.
df.query('pH > 3.0 and pH < 3.5')
df[(df['pH'] > 3.0) & (df['pH'] < 3.5)]  #  Cave: () needed around each condition!!!
df[(df.pH > 3.0) & (df.pH < 3.5)]        #  Cave: () needed around each condition!!!
```

```python
# 7. Further filter the results from 6 to grab only those rows that have a residual sugar less than 2.0. 

df[(df['pH'] > 3.0) & (df['pH'] < 3.5) & (df['residual sugar'] < 2.0)] # Cave: () needed around each condition!!
df.query('pH > 3.0 and pH < 3.5 and `residual sugar` < 2.0')  
# from pandas version 0.25 onwards: can escape column names with whitespace using backticks in query()
# see: https://stackoverflow.com/questions/50697536/pandas-query-function-not-working-with-spaces-in-column-names

# !pip install --upgrade pandas 
# pd.__version__  # '1.3.4'
```

### Part 3 - More Practice

Let's move on to some more complicated things. Use your knowledge of `groupby`s, `sorting`, and the other things that you learned in lecture to answer the following. 

1. Get the average amount of `chlorides` for each `quality` value. 
2. For observations with a `pH` greater than 3.0 and less than 4.0, find the average `alcohol` value by `pH`. 
3. For observations with an `alcohol` value between 9.25 and 9.5, find the highest amount of `residual sugar`. 
4. Create a new column, called `total_acidity`, that is the sum of `fixed acidity` and `volatile acidity`. 
5. Find the average `total_acidity` for each of the `quality` values. 
6. Find the top 5 `density` values. 
7. Find the 10 lowest `sulphates` values. 

```python
import pandas as pd
df = pd.read_csv('data/winequality-white.csv', delimiter =';')
df.head()
```

```python
# 1. Get the average amount of chlorides for each quality value. 
df.groupby('quality').mean().chlorides    # slower, because means for all columns calculated
df.groupby('quality').mean()['chlorides'] # slower, because means for all columns calculated
df.groupby('quality').chlorides.mean()    # faster, because means only for 'chlorides' column calculated
df.groupby('quality')['chlorides'].mean() # faster, because means only for 'chlorides' column calculated
```

```python
timeit df.groupby('quality').mean().chlorides
```

```python
timeit df.groupby('quality').mean()['chlorides']
```

```python
timeit df.groupby('quality').chlorides.mean()
```

```python
timeit df.groupby('quality')['chlorides'].mean()
```

```python
# 2. For observations with a pH greater than 3.0 and less than 4.0, find the average alcohol value by pH. 
df.query('pH > 3.0 and pH < 4.0').groupby('pH').mean().alcohol
df.query('pH > 3.0 and pH < 4.0').groupby('pH').alcohol.mean()
df[(df.pH > 3.0) & (df.pH < 4.0)].groupby('pH').alcohol.mean()
```

```python
timeit df.query('pH > 3.0 and pH < 4.0').groupby('pH').mean().alcohol
```

```python
timeit df.query('pH > 3.0 and pH < 4.0').groupby('pH').alcohol.mean()
```

```python
timeit df[(df.pH > 3.0) & (df.pH < 4.0)].groupby('pH').alcohol.mean()
```

```python
# 3. For observations with an alcohol value between 9.25 and 9.5, find the highest amount of residual sugar. 
df.query('alcohol >= 9.25 and alcohol <= 9.5').max()['residual sugar']
df.query('alcohol >= 9.25 and alcohol <= 9.5')['residual sugar'].max()
df[(df.alcohol >= 9.25) & (df.alcohol <= 9.5)]['residual sugar'].max()
```

```python
# 4. Create a new column, called total_acidity, that is the sum of fixed acidity and volatile acidity.
df['total_acidity'] = df['fixed acidity'] + df['volatile acidity']
df.head()
```

```python
df.eval('total_acidity2 = `fixed acidity` + `volatile acidity`', inplace=True)
df.head()
```

```python
# 5. Find the average total_acidity for each of the quality values.
df.groupby('quality').total_acidity.mean()      # faster (calculates only total_acidity means)
df.groupby('quality')['total_acidity'].mean()   
df.groupby('quality').mean().total_acidity      # slower (calculates all means)
df.groupby('quality').mean()['total_acidity']   
```

```python
timeit df.groupby('quality').total_acidity.mean()
```

```python
timeit df.groupby('quality').mean().total_acidity
```

```python
# 6. Find the top 5 density values.
df.density.sort_values(ascending=False).head()
df.sort_values('density', ascending=False).head()['density']
```

```python
# 7. Find the 10 lowest sulphates values. 
df.sulphates.sort_values().head(10)
df.sort_values('sulphates').head(10)['sulphates']
```

### Part 4 - Practice with Plotting 

1. Plot the average amount of `chlorides` for each `quality` value (1 from Part 3). 
2. Plot the `alcohol` values against `pH` values. Does there appear to be any relationship between the two?
3. Plot `total_acidity` values against `pH` values. Does there appear to be any relationship between the two?
4. Plot a histogram of the `quality` values. Are they evenly distributed within the data set?
5. Plot a boxplot to look at the distribution of `citric acid`. 

```python
import pandas as pd
df = pd.read_csv('data/winequality-white.csv', delimiter =';')
df.head()
```

```python
# 1. Plot the average amount of chlorides for each quality value (1 from Part 3). 
# a) with pandas:
df.groupby('quality').chlorides.mean().plot(kind='bar');
```

```python
# b) with Matplotlib
import matplotlib.pyplot as plt
x = df.groupby('quality').chlorides.mean().index
y = df.groupby('quality').chlorides.mean()
#plt.scatter(x, y);
plt.bar(x, y)
plt.xlabel('quality')
plt.ylabel('chlorides')
plt.show()
```

```python
# c) with seaborn
import seaborn as sns
sns.barplot(x, y); 
```

```python
# d) with plotly
import plotly.graph_objects as go
go.Figure(data=go.Bar(x=x, y=y))  #, mode='markers'))
```

```python
# e) with bokeh
from bokeh.plotting import figure, show, output_notebook
output_notebook()
p = figure()
p.circle(x=x, y=y)
show(p)
```

```python
# 2. Plot the alcohol values against pH values. Does there appear to be any relationship between the two? > No
# a) with pandas
df.plot(x='pH', y='alcohol', kind='scatter', alpha=0.5);
```

```python
# b) with seaborn
sns.scatterplot(x=df.pH, y=df.alcohol, alpha=0.5);
```

```python
# c) with plotly
import plotly.express as px
px.scatter(df, x='pH', y='alcohol', opacity=0.5)
```

```python
# 3. Plot total_acidity values against pH values. Does there appear to be any relationship between the two? 
# > Yes, pH is lower for wines with higher total acidity
# a) with pandas
df['total_acidity'] = df['fixed acidity'] + df['volatile acidity']
df.plot(x='pH', y='total_acidity', kind='scatter', alpha=0.5);
```

```python
# b) with seaborn
sns.scatterplot(x=df.pH, y=df.total_acidity, alpha=0.5);
```

```python
# c) with plotly
import plotly.express as px
px.scatter(df, x='pH', y='total_acidity', opacity=0.5)
```

```python
# 4. Plot a histogram of the quality values. Are they evenly distributed within the data set? 
# > No, they are rather normally distributed
# a) with pandas
df.quality.plot(kind='hist', bins=12)
plt.xlabel('Wine quality');
```

```python
# b) with seaborn
sns.distplot(df['quality'], kde=False, bins=12);

# Plot formatting
plt.title('Histogram')
plt.xlabel('Wine quality')
plt.ylabel('Frequency');
```

```python
# c) with plotly
import plotly.express as px
px.histogram(df, x="quality")
```

```python
# 5. Plot a boxplot to look at the distribution of citric acid.
# a) with pandas
df['citric acid'].plot(kind='box');
```

```python
# b) with seaborn
sns.boxplot(data=df['citric acid'], orient='v');
```

```python
# c) with plotly
import plotly.graph_objects as go

fig = go.Figure()
fig.add_trace(go.Box(y = df['citric acid'],name = 'citric acid'))
fig.show()
```
